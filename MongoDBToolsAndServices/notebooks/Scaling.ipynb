```python
def prep_server(uri):
    cli = pymongo.MongoClient(uri)
    db = cli.admin
    # db.authenticate('test', 'test')
    db = cli.test
    coll = db.stress
    coll.drop()
    print('Prep {}'.format(uri))
    for ix_n in range(1000):
        block = [
            {'ix_n': ix_n, 'ix_b': ix_b}
            for ix_b in range(1000)]
        coll.insert_many(block)
        if ix_n % 40 == 0:
            print()
        print('.', end='', flush=True)
    print()
    coll.create_index([
        ('ix_n', 1),
        ('ix_b', 1)
    ])
    print('Done!')
```

```python
for uri in cluster_uris:
    prep_server(uri)
```
